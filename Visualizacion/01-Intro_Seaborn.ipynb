# Introducción a Seaborn

La visualización de datos es una de las técnicas más utilizada para hacer un EDA.  Seaborn es un paquete basado en Matplotlib pero con instrucciones más amigables.

Nos enfocaremos en lo más básico de Seaborn

## Los Datos

Contexto

Esta base de datos contiene 76 atributos, pero en todas las publicaciones se refieren solamente al uso de 14 de ellos.  En particular, la base de datos de Cleveland es la única base de datos que ha sido utilizada por investigadores en ML, a la fecha.  El campo "target" se refiere a la presencia de enfermedad cardíaca en el paciente.  Es un entero con valores de 0 (no hay) o 1 (si hay).


Columnas
* age         age in years
* sex         (1 = male; 0 = female)
* cp          chest pain type (4 values)
* trestbps    resting blood pressure (in mm Hg on admission to the hospital)
* cholserum   serum cholestoral in mg/dl
* fbs(fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
* restecg     resting electrocardiographic results (values 0, 1, 2)
* thalach     maximum heart rate achieved
* exang       exercise induced angina (1 = yes; 0 = no)
* oldpeak     ST depression induced by exercise relative to rest
* slope       the slope of the peak exercise ST segment
* ca          number of major vessels (0-3) colored by flourosopy
* thal        3 = normal; 6 = fixed defect; 7 = reversable defect
* target      cardiac disease (1 = present; 0 = none)

## Gráficas de Distribución

```python
sns.displot(df['age'],kde=False)
```

## Gráficas de Dispersión

Las gráficas de Dispersión muestran la relación entre dos variable (features) contínuas.

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
df.head()
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df)
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df,hue='sex')
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df,hue='sex',palette='Dark2')
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df,hue='sex',size='age')
```

## Gráficas por Pares (Pairplots)

Las gráficas por Pares realizan gráficas de Dispersión  e histogramas para cada columna en el conjunto de datos. Esto quiere decir que podría ser una gráfica enorme para conjuntos de datos grandes!  Deben utilizarse con caución ya que podría tardarse mucho para generarlas y las imágenes podrían resultar muy pequeñas!

INFO: https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
iris = pd.read_csv('Datos/iris.csv')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
# Mostrar las estimaciones de densidad de los kernels (KDEs por sus siglas en inglés) en vez de histogramas a lo largo de la diagona
sns.pairplot(iris, hue="species")
```
