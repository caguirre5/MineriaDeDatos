# Ejercicios básicos con Seaborn

## Importar librerías

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Importar Datos

```python
tips = sns.load_dataset('tips')
```

## Crear un histograma

```python
sns.histplot(tips['total_bill'])
plt.show()
```

## Crear una gráfica de dispersión

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

## Crear una gráfica de líneas

```python
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.show()
```

## Crear una gráfica de barras

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.show()
```

## Crear una gráfica de caja con bigotes

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

## Crear una gráfica de violín

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

## Crear un mapa de calor

### Ojo! se usan otros datos

```python
flights = sns.load_dataset('flights')
flights = flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights)
plt.show()
```

## Crear gráficas de pares

```python
sns.pairplot(tips)
plt.show()
```

## Crear una gráfica de facetas

```python
g = sns.FacetGrid(tips, col='time')
g.map(sns.histplot, 'total_bill')
plt.show()
```

## Crear una gráfica combinada "jointplot"

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

## Crear una gráfica de categorías en una malla de facetas

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=tips)
plt.show()
```

## Creación de una gráfica de modelo lineal

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```
