## Importar librerías

```python
import seaborn as sns
import seaborn.objects as sbo
import matplotlib.pyplot as plt
import pandas as pd
```

## Importar Datos

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Crear un histograma

```python
(sbo.Plot(data = tips,
          x = "total_bill"

    ).add(sbo.Bar(),sbo.Hist())
)
```

## Crear una gráfica de dispersión

```python
(sbo.Plot(data=tips,
          x = "total_bill",
          y="tip"

    ).add(sbo.Dot())
)
```

## Crear una gráfica de línea

```python
(sbo.Plot(data=tips,
          x = "total_bill",
          y="tip"

    ).add(sbo.Line())
)
```

## Crear una gráfica de barras

```python
(sbo.Plot(data=tips,
          x = "day",
          y="total_bill"

    ).add(sbo.Bar(alpha=.5), sbo.Range())
)
```
