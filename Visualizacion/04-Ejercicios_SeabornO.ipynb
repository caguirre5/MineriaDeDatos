## Importar librer√≠as

```python
import seaborn as sns
import seaborn.objects as sbo
import matplotlib.pyplot as plt
import pandas as pd
```

## Importar Datos

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Crear un histograma

```python
(sbo.Plot(data = tips,
          x = "total_bill"

    ).add(sbo.Bar(),sbo.Hist())
)
```

## Crear una gr√°fica de dispersi√≥n

```python
(sbo.Plot(data=tips,
          x = "total_bill",
          y="tip"

    ).add(sbo.Dot())
)
```

## Crear una gr√°fica de l√≠nea

```python
(sbo.Plot(data=tips,
          x = "total_bill",
          y="tip"

    ).add(sbo.Line())
)
```

## Crear una gr√°fica de barras

```python
(sbo.Plot(data=tips,
          x = "day",
          y="total_bill"

    ).add(sbo.Bar(alpha=.5), sbo.Range())
)
```
